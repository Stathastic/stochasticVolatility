```r
library(readr)

data <- as.data.frame(read_csv("1180.SR.csv"))
date <- data$Date[1:(length(data$Date)-1)]
dates<-as.integer(format(as.Date(date),"%Y%M%d"))
```

```r

data$vol10<-parkinson_volatility_rolling(data, 10)
data$vol30<-parkinson_volatility_rolling(data, 30)
```

```r
# Load necessary libraries
library(ggplot2)

# Ensure that 'Date' is in Date format
data$Date <- as.Date(data$Date)

# Plotting the 10-day rolling volatility
vol_plot<-ggplot(data, aes(x = Date, y = vol10)) + 
    geom_line() +  # Draw the line chart
    geom_point() +  # Add points to each data point
    labs(title = "10-Day Forward Rolling Volatility", 
         x = "Date", 
         y = "Volatility") + 
    theme_minimal()  # Use a minimalistic theme
ggsave("/home/jianj0c/project/STP/Paper/paper_V/65fe88548c149ce7f8f560d6/picture/10_day_rolling_volatility.png", plot = vol_plot, width = 10, height = 6, dpi = 300)
```
